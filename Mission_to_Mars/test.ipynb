```python
import pandas as pd
import requests
from sqlalchemy import create_engine
from splinter import Browser
from bs4 import BeautifulSoup
import time
```

```python
executable_path = {'executable_path': 'c:\chromedriver\chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
time.sleep(2)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
# html_segment = soup.find('div', class_='collapsible results')
print(soup.prettify())
# browser.find_by_css('.itemLink').link.find_by_href('/search/map/Mars/Viking/cerberus_enhanced').click()
browser.find_by_css('.itemLink')
```

```python
itemLink_list = browser.find_by_css('.itemLink')
hemisphere_img_urls = []

for item in itemLink_list:
    if "Enhanced" in item.text:
        hemisphere_img_urls.append({"title": item.text, "img_url": item["href"]})
        # print(i["href"])
for url in hemisphere_img_urls:
    print(url)
url = hemisphere_img_urls[0]["img_url"]
browser.visit(url)
time.sleep(2) 
```
